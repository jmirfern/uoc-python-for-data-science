# Exercici 3 Ús de funcions i fitxers.

<span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px;">EG</span> **(2 punts)**

3.1: (1 punt) Crea una funció que s'anomeni `job_impact_by_year` que rebi com a dades d'entrada:

- La ruta de l'arxiu
- Un any específic.

La funció tornarà un diccionari on la clau sigui l'any i el valor sigui una llista de tuples amb l'impacte laboral: {any : [('categoria', 'xx%'), ('categoria', 'xx%')]}

3.2: (1 punt) Crea una altra funció anomenada `automation_risk_dict` que rebi com a dades d'entrada:

- La ruta de l'arxiu
- Un llindar de risc (numèric).

La funció tornarà un diccionari on la clau és l'any i el valor és una llista de tuples amb els sectors que superen aquest percentatge de risc i el seu valor (Sector, Percentatge): {any : [('sector', 'xx%'), ('sector', 'xx%')]}

```python
# Referències
# Unitat 1, capítol 1, subtítol 1.4 "Altres variants de diccionari"
# Unitat 2, capítol 1 "Ús avançat de funcions"
# Unitat 3, capítol 3 "Lectura i escriptura de fitxers amb Pandas"
# Documentació oficial de Python, mòdul re: https://docs.python.org/3/library/re.html
# HOWTO oficial de regex a Python: https://docs.python.org/3/howto/regex.html

import pandas as pd
import re

def job_impact_by_year(file_path, year):
    """
    Retorna un diccionari amb l'impacte laboral per any especificat.

    La funció carrega el dataframe des del fitxer CSV, filtra la fila corresponent a l'any donat,
    i extreu les dades de les columnes relacionades amb el risc d'automatització de feines.
    El diccionari resultant té l'any com a clau i una llista de tuples (sector, valor) com a valor.

    Args:
        file_path (str): La ruta al fitxer CSV que conté les dades.
        year (int): L'any específic per al qual obtenir l'impacte laboral.

    Returns:
        dict: Un diccionari amb l'any com a clau i una llista de tuples (sector, valor) com a valor.
              Si no es troba l'any, retorna un diccionari buit.
    """
    # Carrega el dataframe des del fitxer CSV
    df = pd.read_csv(file_path)

    # Filtra la fila corresponent a l'any especificat
    row = df[df['Year'] == year]

    # Si no es troba l'any, retorna un diccionari buit
    if row.empty:
        return {}

    sectors = []
    for col in df.columns:
        # Només processa les columnes relacionades amb 
        # l'impacte laboral d'automatització
        if 'Jobs at High Risk of Automation' in col:
            # Extreu el nom del sector de la part de la columna després de " - "
            sector = col.split(' - ')[1].replace(' (%)', '')

            # Agafa el valor corresponent de la fila filtrada
            value = row[col].values[0]

            # Afegeix el tuple (sector, valor) a la llista
            sectors.append((sector, value))
    return {year: sectors}
```

```python
# Referències
# Unitat 1, capítol 1, subtítol 1.4 "Altres variants de diccionari"
# Unitat 2, capítol 1 "Ús avançat de funcions"
# Unitat 3, capítol 3 "Lectura i escriptura de fitxers amb Pandas"



def automation_risk_dict(file_path, threshold):
    """
    Retorna un diccionari amb els sectors que superen un llindar de risc d'automatització per cada any

    La funció carrega el dataframe des del fitxer CSV, itera per cada fila (any),
    i per cada columna relacionada amb el risc d'automatització, comprova si el valor supera el llindar.
    El diccionari resultant té l'any com a clau i una llista de tuples (sector, valor) com a valor.

    Args:
        file_path (str): La ruta al fitxer CSV que conté les dades.
        threshold (float): El llindar numèric de risc (sense '%') per filtrar els sectors.

    Returns:
        dict: Un diccionari amb l'any com a clau i una llista de tuples (sector, valor) com a valor.
              Només inclou anys amb sectors que superen el llindar.
    """

    # Carrega el fitxer CSV en un DataFrame de pandas
    df = pd.read_csv(file_path)

    # Diccionari final on guardar els anys i els sectors seleccionats
    result = {}

    # Itera per cada fila del DataFrame; cada fila representa un any
    for _, row in df.iterrows():
        year = row['Year']
        sectors = []

        # Recorre totes les columnes per identificar les que parlen de risc d'automatització
        for col in df.columns:
            if 'Jobs at High Risk of Automation' in col:
                # Extreu el nom del sector de la columna eliminant informació addicional
                sector = col.split(' - ')[1].replace(' (%)', '')

                # Agafa el valor de la cel·la, que normalment és un string amb '%'
                value_str = row[col]
                value_num = float(value_str.replace('%', ''))

                # Comprova si el valor supera el llindar i, si és així, l'afegeix a la llista
                if value_num > threshold:
                    sectors.append((sector, value_str))

        # Només afegeix l'any al resultat si hi ha sectors que compleixen la condició
        if sectors:
            result[year] = sectors

    return result
```

# Exercici 4: Ús de funcions i expressions regulars

<span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px;">EG</span> **(2 punts)**

4.1 (1 punt) Crea una funció anomenada `filter_automation_columns` que retorni un diccionari. La clau ha de ser el nom de la columna que contingui la paraula "Risk" seguida d'un guió o un espai, i el valor ha de ser la dada corresponent a l'any 2021.
Nota important. Les expressions regulars han de cercar paraules que continguin la paraula "risk" independentment que siguin majúscules o minúscules i ha de ser seguida d'un caràcter de puntuació o un espai en blanc.

4.2 (1 punt) Crea una funció anomenada `get_adoption_metrics` que identifiqui totes les columnes que parlin de "Adoption" o "Implement". La funció ha de tornar un diccionari on la clau sigui el nom de la columna i el valor sigui la dada del darrer any registrat al fitxer. Nota important: Les expressions regulars han de cercar paraules que continguin paraules amb les arrels "Adopt" o "Implement", sent vàlides qualsevol de les dues.

```python
# Referències:
# Unitat 2, capítol 1 "Ús avançat de funcions"
# Unitat 1, capítol 3, subtítol 3.2 "Expressions regulars"
# Unitat 1, capítol 3, subapartat 3.2.1 "Funcions del mòdul re"
# Unitat 1, capítol 3, subapartat 3.2.2 "Sintaxi de les expressions regulars"
# Documentació oficial de Python, mòdul re: https://docs.python.org/3/library/re.html
# Regular Expression HOWTO de Python: https://docs.python.org/3/howto/regex.html


def filter_automation_columns(file_path):
    """
    Retorna un diccionari amb les columnes que contenen 'Risk' seguida d'un guió o espai,
    i el valor corresponent a l'any 2021.

    Utilitza una expressió regular case-insensitive per detectar la paraula 'risk'
    seguida immediatament per un guió o un espai en blanc, i llegeix el valor només
    per l'any 2021.

    Args:
        file_path (str): Ruta al fitxer CSV.

    Returns:
        dict: Diccionari amb nom de columna com a clau i valor de 2021 com a valor.
    """
    # Carrega el fitxer CSV a un DataFrame de pandas
    df = pd.read_csv(file_path)

    # Estableix la columna 'Year' com índex per accedir fàcilment a l'any 2021
    df.set_index('Year', inplace=True)

    result = {}
    # Recorre totes les columnes del DataFrame
    for col in df.columns:
        # Busca la paraula 'risk' seguida d'un guió o un espai, 
        # sense que les majúscules afectin
        if re.search(r'risk[-\s]', col, re.IGNORECASE):
            # Afegeix la columna i el seu valor per a l'any 2021 al diccionari de resultat
            result[col] = df.loc[2021, col]

    return result
```

```python
#test para comprobar el ejercicio 4.1
filter_automation_columns("data/usoDeInteligenciaArtificial.csv")
```

```python
# Referències:
# Unitat 2, capítol 1 "Ús avançat de funcions"
# Unitat 1, capítol 3, subtítol 3.2 "Expressions regulars"
# Unitat 1, capítol 3, subapartat 3.2.1 "Funcions del mòdul re"
# Unitat 1, capítol 3, subapartat 3.2.2 "Sintaxi de les expressions regulars"
# Documentació oficial de Python, mòdul re: https://docs.python.org/3/library/re.html
# Regular Expression HOWTO de Python: https://docs.python.org/3/howto/regex.html

def get_adoption_metrics(file_path):
    """
    Retorna un diccionari amb les columnes que contenen 'Adopt' o 'Implement',
    i el valor del darrer any registrat.
    
    Utilitza expressions regulars per cercar les arrels 'adopt' o 'implement' case-insensitive.
    
    Args:
        file_path (str): Ruta al fitxer CSV.
    
    Returns:
        dict: Diccionari amb nom de columna com a clau i valor del darrer any com a valor.
    """
    # Carrega el fitxer CSV a un DataFrame de pandas
    df = pd.read_csv(file_path)
    
    # Estableix la columna 'Year' com índex per accedir fàcilment als anys
    df.set_index('Year', inplace=True)
    
    # Obté el darrer any registrat al fitxer (el valor màxim de l'índex)
    last_year = df.index.max()
    
    # Inicialitza el diccionari de resultat
    result = {}
    
    # Recorre totes les columnes del DataFrame
    for col in df.columns:
        # Busca les arrels 'adopt' o 'implement' a la columna, 
        # sense que les majúscules afectin (case-insensitive)
        if re.search(r'(adopt|implement)', col, re.IGNORECASE):
            # Afegeix la columna i el seu valor del darrer any al diccionari de resultat
            result[col] = df.loc[last_year, col]
    return result
```

El resultat daquesta funció seran les paraules (que es repeteixen més duna vegada) amb les vegades que es repeteix.

**Notes importants:**

1) Les paraules es compten per si soles, és a dir, si una paraula forma part d'una altra no compta.

2) Si una paraula té alguna lletra majúscula i una altra no, però és la mateixa paraula, es compten com a iguals, per tant, suma al comptador de la paraula.

Exemple:

<pre>
Frase: El menjar de la selva és millor que la lasanya.

Compteig de paraules:

la: 3
menjar: 1
de: 1
selva: 1
és: 1
millor: 1
que: 1
lasanya: 1

Resultat de la funció:

la: 3
</pre>

**Nota:**
Com que es tracta d'un fitxer .docx ha d'utilitzar la llibreria adequada per a la lectura.

```python
# Referències
# Unitat 2, capítol 1 "Ús avançat de funcions"
# Unitat 1, capítol 3, subtítol 3.2 "Expressions regulars"
# Unitat 1, capítol 3, subapartat 3.2.1 "Funcions del mòdul re"
# Unitat 1, capítol 1, subtítol 1.4 "Altres variants de diccionari"
# Documentació oficial de Python, mòdul re: https://docs.python.org/3/library/re.html
# Documentació oficial de collections.Counter: https://docs.python.org/3/library/collections.html#collections.Counter
# Documentació oficial de python-docx: https://python-docx.readthedocs.io/

from docx import Document
import re
from collections import Counter

def lectura(file_path):
    """
    Llegeix un fitxer .docx, compta les paraules i retorna les que es repeteixen més d'una vegada,
    ignorant majúscules/minúscules.
    
    Args:
        file_path (str): Ruta al fitxer .docx
    
    Returns:
        dict: Diccionari amb paraules repetides com a claus i comptador com a valors.
    """
    # Carrega el document .docx des de la ruta especificada
    doc = Document(file_path)  
    # Extreu el text de tots els paràgrafs i els uneix en una sola cadena
    text = ' '.join([p.text for p in doc.paragraphs])  
    # Troba totes les paraules (seqüències de lletres/dígits) en minúscules, ignorant majúscules
    words = re.findall(r'\b\w+\b', text.lower())  
    # Comptabilitza la freqüència de cada paraula utilitzant Counter
    word_counts = Counter(words)  
    # Crea un diccionari només amb paraules que apareixen més d'una vegada
    result = {word: count for word, count in word_counts.items() if count > 1}  
    return result
```
