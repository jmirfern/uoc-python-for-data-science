Justifiqueu breument per què ambdues funcions presenten el mateix ordre de creixement asimptòtic.



- $n^2$ i $n^2 + n$ tenen el mateix ordre de creixement perquè, quan $n$ creix, el terme dominant és $n^2$. El terme lineal $n$ es llavors negligible. Per tant, $n^2 + n \in \Theta(n^2)$.

- $\log(n)$ i $\log(n^2)$ tenen el mateix ordre perquè:
  $$\log(n^2) = 2\log(n)$$
  El factor constant 2 no afecta l’ordre de creixement, així que $\log(n^2) \in \Theta(\log(n))$.

- $n\log(n)$ i $n\log(n^2)$ tenen el mateix ordre perquè:
  $$n\log(n^2) = n \cdot 2\log(n) = 2n\log(n)$$
  Novament, el factor constant 2 no modifica el creixement asimptòtic. Per tant, $n\log(n^2) \in \Theta(n\log(n))$.

- $n$ i $3n$ tenen el mateix ordre perquè només es diferencien per una constant multiplicativa. Així, $3n \in \Theta(n)$.

- $2^n$ i $e^n$ són totes dues funcions de creixement exponencial. En termes generals d’anàlisi d’algorismes, pertanyen al mateix tipus de creixement: exponencial.

- $n!$ i $(n+1)!$ són totes dues funcions de creixement factorial, ja que:
  $$(n+1)! = (n+1)n!$$
  Per tant, totes dues representen un creixement factorial, molt més ràpid que qualsevol funció polinòmica o exponencial simple.

En aquest apartat analitzarem l’eficiència temporal de diferents fragments de codi.

L’objectiu és comprendre com la mida de l’entrada afecta el temps d’execució d’un algorisme i aprendre a identificar possibles ineficiències en la seva implementació.

Recordeu que, en analitzar l’eficiència d’un algorisme, normalment es considera el **pitjor cas**, és a dir, aquell en què l’algorisme requereix el major nombre d’operacions per completar-se.

<span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span> **Exercici 2.1 (0.75 punts)**

Indiqueu la complexitat algorísmica $\Theta$ en relació amb el temps d’execució del fragment de codi següent, justificant la vostra resposta.

````python
def extract_active_bits(x: int) -> int:
    """
    Compute the number of active bits in the binary 
    representation of an int.

    Args:
        x (int): Input integer.

    Returns:
        int: Number of active bits.
    """
    count: int = 0
    value: int = x

    while value > 1:
        bit: int = value % 2
        value = value // 2

        if bit == 1:
            count += 1

    if value == 1:
        count += 1

    return count
````

A cada iteració, value es divideix entre 2 mitjançant divisió entera. Per tant, el nombre d’iteracions no creix linealment amb x, sinó amb el nombre de vegades que podem dividir x entre 2 fins arribar a 1.

Aquest nombre és aproximadament $\log_2(x)$

La resta d’operacions dins del bucle tenen cost constant, i les instruccions finals també són constants. Per això, el temps d’execució ve determinat pel nombre d’iteracions del while.


<span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span> **Exercici 2.2 (0.75 punts)**

En aquest exercici també analitzareu un fragment de codi amb l’objectiu de detectar possibles ineficiències i comprendre la seva complexitat temporal.

Justifiqueu breument l’anàlisi de l’algorisme, indiqueu-ne la complexitat temporal i comenteu on es troba la principal ineficiència del codi.

**Nota:**  
- En aquest exercici podeu assumir que la dimensió dels embeddings és fixa (per exemple, **1408**).

````python
def find_most_similar(query: list[float], embeddings: list[list[float]]) -> int:
    """
    Find the index of the most similar embedding using squared 
    Euclidean distance.

    Args:
        query (list[float]): Query embedding.
        embeddings (list[list[float]]): Database of embeddings.

    Returns:
        int: Index of the most similar embedding.
    """
    best_index = -1
    best_distance = float("inf")

    for i in range(len(embeddings)):
        distance = 0.0

        for j in range(len(query)):
            diff = query[j] - embeddings[i][j]
            distance += diff * diff

        if distance < best_distance:
            best_distance = distance
            best_index = i

    return best_index
````

La complexitat temporal de la funció és:

$\Theta(n)$ 

on $n$ és el nombre d’embeddings.

La funció recorre tots els embeddings de la base de dades i, per a cadascun, calcula la distància euclidiana quadrada respecte de l’embedding de consulta. Aquest càlcul fa servir un bucle intern que recorre totes les components del vector. Ara bé, com que l’enunciat indica que la dimensió dels embeddings és fixa, aquest cost es pot considerar constant. Per tant, el temps d’execució creix linealment amb el nombre d’embeddings.

La complexitat general seria:

$(\Theta(n \cdot d))$

on $d$ és la dimensió de cada embedding. 

Però si $d$ és fixa, queda:

$(\Theta(n))$

Utilitzant les eines de *profiling* que hem vist al Notebook de teoria, analitzeu el fragment de codi ineficient següent i identifiqueu quin és el coll d’ampolla de l’algorisme.

Per fer-ho, heu de dur a terme el següent:

- Genereu una llista d’enters (10.000 elements) per poder analitzar el rendiment de l’algorisme.

- Analitzeu el rendiment de la funció utilitzant `%timeit` i `%lprun`.

- A partir dels resultats obtinguts amb ambdues eines, identifiqueu quina part de l’algorisme constitueix el coll d’ampolla i expliqueu les ineficiències detectades, relacionant-les amb els resultats obtinguts mitjançant el *profiling*.

- Indiqueu quina és la complexitat temporal aproximada de l’algorisme i justifiqueu la vostra resposta.

````python
from typing import List, Dict

def compute_frecs(values: List[int]) -> Dict[int, int]:
    """
    Computes the frequency of each value in the input list.

    Args:
        values: List of integer values.

    Returns:
        Dictionary mapping each value to its frequency.
    """
    frecs: Dict[int, int] = {}
    for j in range(len(values)):
        current_value = values[j]
        count = 0
        for k in range(len(values)):
            if values[k] == current_value:
                count += 1
        frecs[current_value] = count
    return frecs
````

````python
import random

# Generem una llista de 10.000 números aleatoris compresos entre 0 i 1000
values = [random.randint(0, 1000) for _ in range(10000)]

# Apliquem la funció compute_frecs
%timeit compute_frecs(values)

# Analitzem el rendiment mitjançant el mòdul line_profiler aplicant la llista anterior sobre la funció d'estudi
%load_ext line_profiler
%lprun -f compute_frecs compute_frecs(values)
````

Els resultats de profiling mostren que el coll d’ampolla principal es troba en el bucle intern de la funció compute_frecs, concretament en aquesta part:

```python
for k in range(len(values)):
    if values[k] == current_value:
        count += 1
```

Per cada element de la llista, l’algorisme torna a recórrer tota la llista per comptar quantes vegades apareix aquell valor. Això provoca moltes comparacions repetides, especialment quan hi ha valors duplicats, ja que es recalcula la freqüència d’un mateix nombre diverses vegades.

Amb %timeit es pot observar que el temps d’execució és elevat per a una llista de 10.000 elements. Amb %lprun es pot veure que la major part del temps es concentra dins del segon bucle, on es fan les comparacions ```values[k] == current_value.```

La complexitat temporal aproximada de l’algorisme és Θ(n²), on _n_ és la mida de la llista. Això és així perquè hi ha un bucle extern que recorre els n elements i, per cada iteració, un bucle intern que torna a recórrer els _n_ elements. Per tant, el nombre total d’operacions creix aproximadament com $ n · n = n^2$.


<span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span> **Exercici 3.2 (0.5 punts)**

En aquest exercici heu de millorar l’algorisme `compute_frecs` amb l’objectiu de reduir-ne el temps d’execució i millorar-ne l’eficiència.

Per fer-ho, creeu una nova funció anomenada `compute_frecs_improved` i implementeu el codi corresponent.

A més, heu de:
- Mesureu el temps d’execució de `compute_frecs_improved` utilitzant `%timeit`, compareu-lo amb l’obtingut anteriorment per a `compute_frecs` i calculeu l’*speed-up* de la versió optimitzada respecte de l’original.
- Expliqueu quina millora heu introduït respecte de la versió original de l’algorisme.
- Indiqueu quina és la complexitat temporal aproximada de la nova versió.

````python
from typing import List, Dict

def compute_frecs_improved(values: List[int]) -> Dict[int, int]:
    """
    Computes the frequency of each value in the input list.

    Args:
        values: List of integer values.

    Returns:
        Dictionary mapping each value to its frequency.
    """
    frecs: Dict[int, int] = {}

    for value in values:
        if value in frecs:
            frecs[value] += 1
        else:
            frecs[value] = 1

    return frecs
````

````python
import timeit

# Mesurem el temps d'execució de la funció original
temps_original = timeit.timeit(
    lambda: compute_frecs(values),
    number=10
)

# Mesurem el temps d'execució de la funció millorada
temps_millorat = timeit.timeit(
    lambda: compute_frecs_improved(values),
    number=10
)

# Calculem el rati de millora entre les dues mesures anteriors i imprimim resultats
speed_up = temps_original / temps_millorat

print(f"Temps original: {temps_original:.6f} s")
print(f"Temps millorat: {temps_millorat:.6f} s")
print(f"Speed-up: {speed_up:.2f}x")
````

La millora introduïda consisteix a comptar les freqüències en una sola passada per la llista. En la versió original, per cada element es tornava a recórrer tota la llista per comptar quantes vegades apareixia, cosa que provocava moltes comparacions repetides.

En canvi, la funció compute_frecs_improved utilitza un diccionari per anar acumulant el nombre d’aparicions de cada valor. Si el valor ja existeix al diccionari, s’incrementa el seu comptador; si no existeix, s’inicialitza a 1.

La complexitat temporal aproximada de la nova versió és Θ(n), on n és el nombre d’elements de la llista. Això és així perquè la funció només recorre la llista una vegada i les operacions d’accés i actualització del diccionari tenen cost constant de mitjana.

Per tant, respecte de la versió original, que tenia complexitat Θ(n²), aquesta versió és molt més eficient.

Es proporciona una funció que, donada una imatge en escala de grisos representada mitjançant una matriu, compta quants píxels tenen un valor superior a un determinat llindar. A partir d’aquesta funció ja implementada, es demana desenvolupar dues noves funcions:

- **Versió seqüencial:** implementeu una funció que rebi una llista d’imatges i apliqui seqüencialment la funció `count_white_pixels()`, mesurant i mostrant el temps total d’execució en segons.

- **Versió amb multiprocessament:** implementeu una altra funció que realitzi el mateix càlcul, però utilitzant múltiples processos mitjançant el mòdul `multiprocessing`, mostrant també el temps total d’execució.

Ambdues funcions hauran d’utilitzar la mateixa llista d’entrada i permetre comparar el rendiment entre l’execució seqüencial i la paral·lela. Per fer-ho, també es proporciona el llistat de matrius, que simulen imatges de diferents mides i que s’haurà d’utilitzar com a entrada comuna en ambdues implementacions.

**IMPORTANT**: depenent del sistema i de l’entorn d’execució, el multiprocessament pot no funcionar correctament directament en un notebook. Si això passa, hauràs de moure les funcions a un mòdul de Python independent (.py).

````python
import numpy as np
import multiprocessing as mp

def count_white_pixels(image: np.ndarray, threshold: int = 200) -> int:
    """
    Counts how many pixels in a grayscale image are greater than a 
    given threshold.

    Args:
        image (np.ndarray): Image represented as a 2D NumPy array.
        threshold (int): Threshold value used to determine whether 
        a pixel is considered white.

    Returns:
        int: Number of pixels greater than the threshold.
    """
    count = 0
    rows, cols = image.shape

    for i in range(rows):
        for j in range(cols):
            if image[i, j] > threshold:
                count += 1

    return count


images = [
    np.random.randint(0, 256, (2000, 2000), dtype=np.uint8),
    np.random.randint(0, 256, (2500, 2500), dtype=np.uint8),
    np.random.randint(0, 256, (3000, 3000), dtype=np.uint8),
    np.random.randint(0, 256, (3500, 3500), dtype=np.uint8),
]
````

````python
import time
import multiprocessing as mp
import os

# Generem una funció que aplica la funció `count_white_pixels` de 
# forma seqüencial, és a dir, un per un, i compta el temps total

def count_white_pixels_sequential(images):
    start_time = time.perf_counter()

    results = [count_white_pixels(image) for image in images]

    total_time = time.perf_counter() - start_time
    print(f"Temps seqüencial: {total_time:.4f} segons")

    return results

# Generem una funció que aplica la funció `count_white_pixels` de 
# forma paral·lela i compta el temps total

def count_white_pixels_parallel(images):
    start_time = time.perf_counter()

    num_processes = min(len(images), os.cpu_count())

    ctx = mp.get_context("fork")

    with ctx.Pool(processes=num_processes) as pool:
        results = pool.map(count_white_pixels, images)

    total_time = time.perf_counter() - start_time
    print(f"Temps paral·lel: {total_time:.4f} segons")

    return results

````

````python
# Verifiquem que ambdues funcions produeixen el mateix resultat

results_seq = count_white_pixels_sequential(images)
results_par = count_white_pixels_parallel(images)

assert results_seq == results_par

print("Les dues versions produeixen el mateix resultat.")
````

````python
import json
import requests

def download_json(url: str, output_path: str) -> None:
    """
    Downloads a JSON resource from a URL and saves it locally.

    Args:
        url: URL of the JSON resource.
        output_path: Local path where the JSON file will be stored.
    """
    response: requests.Response = requests.get(url, timeout=10)
    response.raise_for_status()

    data: dict = response.json()

    # Save JSON content into a local file
    with open(output_path, "w", encoding="utf-8") as file:
        json.dump(data, file, indent=4, ensure_ascii=False)
        
````

Durant el procés, heu de:

````python
def download_products_sequential(num_products: int = 75, output_dir: str = "products") -> int:
    """
    Descarrega els JSON files de forma sequencial i els emmagatzema localment

    Args:
        num_products: Nombre d'arxius JSON de productes a descarregar
        output_dir: Directori on es guardaran els fitxers

    Returns:
        int: Nombre d'arxius JSON de productes descarregats correctament
    """

    # En cas de que no existeixi el directori, el creem
    os.makedirs(output_dir, exist_ok=True)

    urls = get_json_urls(num_products)
    downloaded = 0

    start_time = time.perf_counter()

    for i, url in enumerate(urls, start=1):
        output_path = os.path.join(output_dir, f"product_{i}.json")

        try:
            download_json(url, output_path)
            downloaded += 1
        except requests.RequestException as error:
            print(f"Error descarregant {url}: {error}")

    total_time = time.perf_counter() - start_time

    print(f"Temps total d'execució seqüencial: {total_time:.4f} segons")
    print(f"Productes descarregats correctament: {downloaded}")

    return downloaded
````

Implementeu una versió concurrent del sistema de descàrrega de productes.

Per fer-ho, heu de reutilitzar la lògica de descàrrega de l’exercici anterior i adaptar-la perquè múltiples productes es puguin descarregar de manera concurrent mitjançant la classe `ThreadPoolExecutor`.

Cada fitxer s’haurà de continuar emmagatzemant localment a la carpeta `products/`, utilitzant un nom únic per a cada producte.

Per tant:

- Compareu breument el resultat obtingut amb el temps de la versió seqüencial.
- Proveu diferents valors per al nombre de threads amb la finalitat de determinar quin ofereix el millor rendiment.
- Representeu en una gràfica els temps obtinguts per als diferents valors de threads.
- Indiqueu quin nombre de threads proporciona el millor resultat.

**Nota:**

Podeu provar, per exemple, valors entre 1 i 10 per a `max_workers`.

````python
import os
import time
import requests
import matplotlib.pyplot as plt
from concurrent.futures import ThreadPoolExecutor, as_completed

def download_one_product(index_url: tuple[int, str], output_dir: str = "products") -> bool:
    """
    Downloads a single product JSON file.

    Args:
        index_url: Tuple containing the product index and URL.
        output_dir: Directory where the JSON file will be saved.

    Returns:
        bool: True if the download was successful, False otherwise.
    """
    index, url = index_url
    output_path = os.path.join(output_dir, f"product_{index}.json")

    try:
        download_json(url, output_path)
        return True
    except requests.RequestException as error:
        print(f"Error descarregant {url}: {error}")
        return False

def download_products_concurrent(
    num_products: int = 75,
    output_dir: str = "products",
    max_workers: int = 5
) -> tuple[int, float]:
    """
    Downloads product JSON files concurrently using ThreadPoolExecutor.

    Args:
        num_products: Number of product JSON files to download.
        output_dir: Directory where the JSON files will be saved.
        max_workers: Number of threads used by ThreadPoolExecutor.

    Returns:
        tuple[int, float]: Number of successful downloads and total execution time.
    """
    os.makedirs(output_dir, exist_ok=True)

    urls = get_json_urls(num_products)
    indexed_urls = list(enumerate(urls, start=1))

    start_time = time.perf_counter()

    downloaded = 0

    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures = [
            executor.submit(download_one_product, index_url, output_dir)
            for index_url in indexed_urls
        ]

        for future in as_completed(futures):
            if future.result():
                downloaded += 1

    total_time = time.perf_counter() - start_time

    print(f"Threads: {max_workers}")
    print(f"Temps concurrent: {total_time:.4f} segons")
    print(f"Productes descarregats correctament: {downloaded}")

    return downloaded, total_time
````

````python
thread_values = list(range(1, 11))
times = []
downloads = []

for workers in thread_values:
    downloaded, total_time = download_products_concurrent(max_workers=workers)
    downloads.append(downloaded)
    times.append(total_time)

plt.figure(figsize=(8, 5))
plt.plot(thread_values, times, marker="o")

plt.xlabel("Nombre de threads")
plt.ylabel("Temps d'execució (segons)")
plt.title("Temps de descàrrega segons el nombre de threads")
plt.grid(True)
plt.show()

best_index = times.index(min(times))
best_threads = thread_values[best_index]
best_time = times[best_index]

print(f"Millor nombre de threads: {best_threads}")
print(f"Millor temps: {best_time:.4f} segons")
````
