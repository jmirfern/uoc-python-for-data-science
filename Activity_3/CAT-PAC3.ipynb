````python
def extract_active_bits(x: int) -> int:
    """
    Compute the number of active bits in the binary 
    representation of an int.

    Args:
        x (int): Input integer.

    Returns:
        int: Number of active bits.
    """
    count: int = 0
    value: int = x

    while value > 1:
        bit: int = value % 2
        value = value // 2

        if bit == 1:
            count += 1

    if value == 1:
        count += 1

    return count
````

<span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span> **Exercici 2.2 (0.75 punts)**

En aquest exercici també analitzareu un fragment de codi amb l’objectiu de detectar possibles ineficiències i comprendre la seva complexitat temporal.

Justifiqueu breument l’anàlisi de l’algorisme, indiqueu-ne la complexitat temporal i comenteu on es troba la principal ineficiència del codi.

**Nota:**  
- En aquest exercici podeu assumir que la dimensió dels embeddings és fixa (per exemple, **1408**).

````python
def find_most_similar(query: list[float], embeddings: list[list[float]]) -> int:
    """
    Find the index of the most similar embedding using squared 
    Euclidean distance.

    Args:
        query (list[float]): Query embedding.
        embeddings (list[list[float]]): Database of embeddings.

    Returns:
        int: Index of the most similar embedding.
    """
    best_index = -1
    best_distance = float("inf")

    for i in range(len(embeddings)):
        distance = 0.0

        for j in range(len(query)):
            diff = query[j] - embeddings[i][j]
            distance += diff * diff

        if distance < best_distance:
            best_distance = distance
            best_index = i

    return best_index
````

````python
from typing import List, Dict

def compute_frecs(values: List[int]) -> Dict[int, int]:
    """
    Computes the frequency of each value in the input list.

    Args:
        values: List of integer values.

    Returns:
        Dictionary mapping each value to its frequency.
    """
    frecs: Dict[int, int] = {}
    for j in range(len(values)):
        current_value = values[j]
        count = 0
        for k in range(len(values)):
            if values[k] == current_value:
                count += 1
        frecs[current_value] = count
    return frecs
````

````python
import json
import requests

def download_json(url: str, output_path: str) -> None:
    """
    Downloads a JSON resource from a URL and saves it locally.

    Args:
        url: URL of the JSON resource.
        output_path: Local path where the JSON file will be stored.
    """
    response: requests.Response = requests.get(url, timeout=10)
    response.raise_for_status()

    data: dict = response.json()

    # Save JSON content into a local file
    with open(output_path, "w", encoding="utf-8") as file:
        json.dump(data, file, indent=4, ensure_ascii=False)
        
````
